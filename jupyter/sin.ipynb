```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols('x')
expr = sp.sin(x)

f = sp.lambdify(x, expr, modules=['numpy'])
x_vals = np.linspace(-np.pi, np.pi, 100)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of sin(x)')
plt.grid(True)
plt.show()
```
